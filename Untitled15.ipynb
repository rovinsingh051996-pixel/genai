```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.output_parsers import StrOutputParser

import os
from getpass import getpass
os.environ['GROQ_API_KEY'] = getpass('рџ”‘ Enter Groq API Key: ')
print('вњ… API Key set!')

# вњ… llama-3.1-8b-instant = official replacement for deprecated llama3-8b-8192
llm = ChatGroq(model='llama-3.1-8b-instant', temperature=0.7, max_tokens=1024)

chat_prompt = ChatPromptTemplate.from_messages([
    ('system', 'You are a helpful AI assistant. Be clear and concise.'),
    MessagesPlaceholder(variable_name='history'),
    ('human', '{input}'),
])
chain = chat_prompt | llm | StrOutputParser()

chat_history = []

def chat(user_input):
    response = chain.invoke({'input': user_input, 'history': chat_history})
    chat_history.append(HumanMessage(content=user_input))
    chat_history.append(AIMessage(content=response))
    return response

query = input("рџ‘¤ Enter queries (comma separated): ").split(",")
query = [q.strip() for q in query]

for q in query:
    print(f'\nрџ‘¤ You: {q}')
    print(f'рџ¤– Bot: {chat(q)}')

print(f'\nвњ… Memory: {len(chat_history)} messages stored')
```
